```python
df = pd.read_csv('adult.data',header = -1)
```

```python
df.columns = ['Age','Workclass','Fnlwgt','Education','Education-Num',
              'Marital_status','Occupation','Relationship','Race','Sex',
              'Capital-Gain','Capital-loss','HPW','Country','Wage']

df
```

```python
df = df.iloc[:10000,:]
```

```python
df['Wage'] = df['Wage'].replace({' <=50K':1,' >50K':0})
```

```python
X = df.drop(['Wage'],axis = 1)
Y = df['Wage']
```

```python
print(X['Country'].value_counts().sort_values(ascending=False).head())
```

```python
X['Country'] = ['United-States' if i ==' United-States' else 'Other' for i in X['Country']]
```

```python
X['Sex'] = X['Sex'].replace({' Male':0,' Female':1})
```

```python
for col in X.columns:
    if X[col].dtypes == 'object':
        unique_cat = len(X[col].unique())
        print(col,unique_cat)
```

```python
dummy =['Workclass','Education','Marital_status','Occupation','Relationship','Race','Country','Sex']
```

```python
def create_dummy(df,dumb):
    for x in dumb:
        dummies = pd.get_dummies(df[x],prefix = x,dummy_na = False)
        df = df.drop(x,1)
        df = pd.concat([df,dummies],axis=1)
    return df
```

```python
X = create_dummy(X,dummy)
```

```python
from itertools import combinations
from sklearn.preprocessing import PolynomialFeatures
```

```python
def interaction(df):
    com = list(combinations(list(df.columns),2))
    colname = list(df.columns) + ['_'.join(x) for x in com]
    
    poly = PolynomialFeatures(interaction_only=True,include_bias = False)
    df = poly.fit_transform(df)
    df = pd.DataFrame(df)
    df.columns = colname
    newindex = [i for i,x in enumerate(list((df == 0).all()))if x]
    df = df.drop(df.columns[newindex],axis=1)
    return df
```

```python
X = interaction(X)
X.head()
```

```python
x,xt,y,yt = train_test_split(X,Y,test_size=.2,random_state=1)
```

```python
import sklearn.feature_selection
```

```python
select = sklearn.feature_selection.SelectKBest(k=20)
```

```python
selfe = select.fit(x,y)
```

```python

index = selfe.get_support(indices=True)
selcol = [X.columns[i] for i in index]

x_sel = x[selcol]
xt_sel = xt[selcol]
```

```python
from sklearn.metrics import accuracy_score
lr = LogisticRegression()
lr.fit(x_sel,y)
pred = lr.predict(xt_sel)
accuracy_score(yt,pred)
```
